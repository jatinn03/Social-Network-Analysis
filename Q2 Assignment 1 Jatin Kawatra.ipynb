```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with weighted edges
G = nx.Graph()

# Add nodes
nodes = [1, 2, 3, 4, 5]
G.add_nodes_from(nodes)

# Add edges with weights
edges = [(1, 2, 5), (1, 3, 2), (2, 4, 3), (2, 5, 7), (3, 4, 4), (4, 5, 1), (3, 5, 6)]
G.add_weighted_edges_from(edges)

# Calculate and print degree centrality
degree_centrality = nx.degree_centrality(G)
print("Degree Centrality:", degree_centrality)

# Calculate and print weighted degree centrality
weighted_degree_centrality = {node: sum(data['weight'] for _, _, data in G.edges(node, data=True)) for node in G.nodes()}
print("Weighted Degree Centrality:", weighted_degree_centrality)

# Visualize the graph
pos = nx.spring_layout(G)
weights = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", font_weight="bold")
nx.draw_networkx_edge_labels(G, pos, edge_labels=weights)
nx.draw_networkx_edges(G, pos, width=[weight for weight in weights.values()])

plt.title("Graph with Weighted Edges")
plt.show()
```
