```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph with two disconnected components
H = nx.Graph()

# Add nodes and edges for the first component
component1 = [1, 2, 3, 4]
edges1 = [(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)]
H.add_nodes_from(component1)
H.add_edges_from(edges1)

# Add nodes and edges for the second component
component2 = [5, 6, 7, 8]
edges2 = [(5, 6), (6, 7), (7, 8), (8, 5), (5, 7)]
H.add_nodes_from(component2)
H.add_edges_from(edges2)

# Calculate connected components and their sizes
connected_components = list(nx.connected_components(H))
component_sizes = [len(comp) for comp in connected_components]
print("Connected Components:", connected_components)
print("Component Sizes:", component_sizes)

# Visualize the graph with different colors for components
colors = ['red', 'blue']
node_colors = []
for node in H:
    for i, comp in enumerate(connected_components):
        if node in comp:
            node_colors.append(colors[i])

pos = nx.spring_layout(H)
nx.draw(H, pos, with_labels=True, node_size=700, node_color=node_colors, font_weight="bold")

plt.title("Graph with Multiple Components")
plt.show()
```
