```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
DG = nx.DiGraph()

# Add nodes
nodes = [1, 2, 3, 4, 5]
DG.add_nodes_from(nodes)

# Add directed edges
edges = [(1, 2), (2, 3), (3, 1), (4, 1), (4, 2), (5, 4), (3, 5), (5, 2)]
DG.add_edges_from(edges)

# Calculate and print in-degree and out-degree
in_degrees = dict(DG.in_degree())
out_degrees = dict(DG.out_degree())
print("In-Degree:", in_degrees)
print("Out-Degree:", out_degrees)

# Visualize the directed graph
pos = nx.spring_layout(DG)
nx.draw(DG, pos, with_labels=True, node_size=700, node_color="lightgreen", font_weight="bold", arrowsize=20)
nx.draw_networkx_edges(DG, pos, arrowstyle='-|>', arrowsize=15)

plt.title("Directed Graph")
plt.show()
```
